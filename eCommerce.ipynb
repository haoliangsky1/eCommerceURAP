```python
import pandas as pd
from html.parser import HTMLParser
from bs4 import BeautifulSoup
import re
```

```python
df = pd.read_csv('test.csv', encoding = "ISO-8859-1")
df = df.head(83)
users = df['user_url']
timelines = df['timeline_html']
likes = df['likes_html']
friends = df['friends_html']
```

Extract each user’s post content and the number of likes from the column timeline_html. Generate a new table: the first column is the user URL and each following column only contains one post content and the number of its likes. This table should have clear timeline information of all 83 users, and each user’s information is in one row.  


```python
post = []
likeList = []
for i in range(0, len(users)):
    user = users[i]
    timeline = timelines[i]
    if type(timeline) is str:
        soup = BeautifulSoup(timeline, 'html.parser')
        text = [soup.get_text()]
        post.append(text)
        likeTag = soup.find_all("a", attrs={"class": "_27jf _3emk"})
        if len(likeTag) != 0:
            likeTag = likeTag[0]['aria-label']
            temp = [int(s) for s in likeTag.split() if s.isdigit()]
            temp1 = ''.join(str(e) for e in temp)
            if temp1 is not '':
                likeList.append(int(temp1))
            else:
                likeList.append(0)
        else:
            likeList.append(0)
    else:
        post.append([])
        likeList.append(0)
len(post)
```

```python
col = ['user'] + ['post'] + ['likes']
df1 = pd.DataFrame(index = range(0, len(users)), columns = col)
for i in range(0, len(users)):
    df1.iloc[i, 0] = users[i]
    df1.iloc[i, 1] = ''.join(post[i])
    df1.iloc[i, 2] = likeList[i]
df1.head()
```

Extract each user’s liked public page name from the column likes_html. Generate a new table: the first column is the user URL and each following column only contains one user’s liked public page. This table should have clear information on all liked pages from 83 users, and each user’s information is in one row.  


```python
publicPage = []
for i in range(0, len(users)):
#     print(i)
    user = users[i]
    like = likes[i]
    currentUser = []
    if type(like) is str:
        soup = BeautifulSoup(like, 'html.parser')
        tagSet = soup.find_all(attrs={"data-tooltip-content": "Verified PageFacebook confirmed this is an authentic Page for this public figure, media company or brand."})
        for tag in tagSet:
            a = tag.previousSibling
            name = a.get_text()
            website = a['href']
            combo = str(name) + ' : ' + str(website)
            currentUser.append(combo)
#         print(currentUser)
        publicPage.append(currentUser)
    else:
        publicPage.append([])
len(publicPage)
maxLen = 0
for l in publicPage:
    if len(l) > maxLen:
        maxLen = len(l)
```

```python
col = ['user'] + ['LikedPages']*maxLen
df2 = pd.DataFrame(index = range(0, len(users)), columns = col)
# df2.head()
for i in range(0, len(users)):
    df2.iloc[i, 0] = users[i]
    if len(publicPage[i]) != 0:
        for j in range(0, len(publicPage[i])):
            df2.iloc[i, j+1] = publicPage[i][j]
df2.head()
```

3.	Extract each user’s friends names from the column friends_html. Generate a new table: the first column is the user URL and each following column only contains one friend page name. This table should have clear information on all liked pages from 83 users, and each user’s information is in one row.  


```python
friendPage = []
for i in range(0, len(users)):
#     print(i)
    user = users[i]
    friend = friends[i]
    currentUser = []
    if type(friend) is str:
        soup = BeautifulSoup(friend, 'html.parser')
        tagSet = soup.find_all(attrs = {"class" : "fsl fwb fcb"})
        for tag in tagSet:
            if tag.contents != []:
                a = tag.contents[0]
                website = a['href']
                name = a.get_text()
            combo = str(name) + " : " + str(website)
            currentUser.append(combo)
        friendPage.append(currentUser)
    else:
        friendPage.append([])
maxLen = 0
for l in friendPage:
    if len(l) > maxLen:
        maxLen = len(l)
print(maxLen)
```

```python
col = ['user'] + ['Friends']*maxLen
df3 = pd.DataFrame(index = range(0, len(users)), columns = col)
# df2.head()
for i in range(0, len(users)):
    df3.iloc[i, 0] = users[i]
    if len(friendPage[i]) != 0:
        for j in range(0, len(friendPage[i])):
            df3.iloc[i, j+1] = friendPage[i][j]
df3.head()
```

4.	Join all the above tables in one table. The key for joining is each users’ URL. 


```python
temp = pd.merge(df1, df2, on = 'user', how = 'outer')
finalTable = pd.merge(temp, df3, on = 'user', how = 'outer')
finalTable.head()
# finalTable = df1.merge(df2.merge(df3, left_on = 'user'), left_on = 'user')
```

Output the tables as csv files

```python
df1.to_csv('Q1.csv')
df2.to_csv('Q2.csv')
df3.to_csv('Q3.csv')
finalTable.to_csv('Q4.csv')
```
